```python
def get_url(position,location):
    template ='https://in.indeed.com/jobs?q={}&l={}+Maharashtra&ts=1733471415842&pts=1733400145270&from=searchOnHP&rq=1&rsIdx=0&fromage=las'
    url=template.format(position,location)
    return url

```

### Lets Extract the Whole page 

```python
def get_record(card):
    atag= card.h2.a.span
    job_title= atag.get('title')
    a_tag_url= card.h2.a
    job_url = 'http://indeed.com' + a_tag_url.get('href')
    company_name =card.find('span','css-1h7lukg eu4oa1w0').text
    job_location= card.find('div','css-1restlb eu4oa1w0').text
    salary= card.find('div', 'css-18z4q2i eu4oa1w0')
    
    if salary:
        salary = salary.text.strip()
        
        # If salary contains "Full time", replace it with "Not Provided"
        if "Full-time" in salary:
            salary = "Not Provided"
    else:
        salary = "Not Provided"
   

    record= (job_title,job_url,company_name,job_location,salary)

    return record

    

    
    
    
    
```

```python
records =[]

for card in cards:
    record= get_record(card)
    records.append(record)
```
